```python
from crewai import Task, Crew, Agent
from textwrap import dedent
import os, json
os.environ["OPENAI_API_KEY"] = input("Вставь свой OpenAI API ключ: ")
```

```python
def load_agents():
    base_path = "agents"
    agent_files = ["project_manager.json", "business_analyst.json", "ui_ux_designer.json", "architect.json"]
    agents = []
    for file in agent_files:
        with open(f"{base_path}/{file}", "r") as f:
            config = json.load(f)
            agent = Agent(
                role=config["role"],
                goal=config["goal"],
                backstory=config["backstory"],
                verbose=True
            )
            agents.append(agent)
    return agents
```

```python
agents = load_agents()

task = Task(
    description=dedent("""
        You are tasked with creating a project plan for developing a website
        with a personal user account system. Include development stages,
        tech stack, architecture, and UX considerations.
    """),
    agent=agents[0]  # Project Manager
)

crew = Crew(agents=agents, tasks=[task], verbose=True)
result = crew.kickoff()

print(result)
```
